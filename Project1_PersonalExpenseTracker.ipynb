```python

#Description

#In todayâ€™s fast-paced world, individuals need to track and manage their expenses effectively. Your task is to build a personal expense tracker that allows users to log daily expenses, categorize them, and track spending against a monthly budget. The tracker should also be able to save and load expenses from a file for future reference.

#Objectives:

#1. Design and implement a personal expense tracker that enables users to manage their expenses

#2. Allow users to categorize expenses and set monthly budgets

#3. It Implement file-handling functionality to save and load expense data

#4.  We Create an interactive, menu-driven interface for ease of use

'''
#This Expense tracker Take 4 inputs date, category, amount and description and using the switch case.
It check a valid date and takes data
'''
```

```python
import os
from datetime import datetime

def add_expense():
    #Add New Expense
    date=input('Enter the current date (YYYY-MM-DD):')
    try:
        date_object = datetime.strptime(date, "%Y-%m-%d")
        print("Valid date:", date_object.date())
    except ValueError:
        print("Invalid date format! Please enter in YYYY-MM-DD format."  )
        add_expense() 
    category=input('Enter the Expense Category:')
    amount=input('Enter the amount:')
    desc=input('Enter the Description:')
    save_expense(date,category,amount,desc)
    print("Expense added successfully!\n")
    
def del_expense():
    os.remove("my_expense.csv")
    
def view_expense():
    print("Date          | Category       | Amount  | Description")
    print("-" * 40)
    fp = open('my_expense.csv', 'r')
    for line in fp:
        date, category, amount, desc = line.strip().split(",")
        print("{}          | {}           | rs {} | {}".format(date, category, amount, desc))
        

def save_expense(date,category,amount,desc):
    fp = open('my_expense.csv', 'a+')
    detail = date + ',' + category + ',' + amount  + ',' + desc + '\n'
    fp.write(detail)
    fp.close()
    
def track_budget():
    amt = 0.0
    """Track spending against a monthly budget."""
    budget = float(input("Enter your monthly budget: "))

    # Load the file
    fp = open('my_expense.csv', 'r')
    for line in fp:
        date, category, amount, desc = line.strip().split(",")
        amt = amt + float(amount)
    remaining_budget = budget - amt
    
    print(f"Total spent: rs{amt:.2f}")
    print(f"Remaining budget: rs{remaining_budget:.2f}\n")
    if(remaining_budget<0):
        print('You have excedded your monthly budget')

def main():
    # expenses = load_expenses()
    
    while True:
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Track Budget")
        print("4. Exit")
        
        choice = input("Choose an option: ")
        
        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expense()
        elif choice == "3":
            track_budget()
        elif choice == "4":
            print("Exiting... Have a great day!")
            break
        else:
            print("Invalid choice! Please try again.\n")


main()
    
```
